### Exercícios Práticos para os Métodos Iterativos de Solução de Sistemas de Equações Lineares

1. Considere os sistemas de equações exemplo apresentados em aula para cada método abordado (constam no PDF da aula). Utilize o código desenvolvido de cada método para resolver, de forma numérica, o respectivo problema.

Para isso, considere, em cada problema, uma condição inicial nula $x^{(0)} = 0$, e como critérios de parada um erro de $\varepsilon = 10^{-2}$ e um número máximo de iterações $k_{\text{max}} = 10$.

```python
import numpy as np
```

#### Método de Jacobi-Richardson

```python
A = np.array([
    [10.0, 2.0, 1.0],
    [1.0, 5.0, 1.0],
    [2.0, 3.0, 10.0]
])
b = np.array([7.0, -8.0, 6.0])
itMax = 10
epslon = 1e-2
n = A.shape[0]

L = np.zeros((n, n))
D = np.zeros((n, n))
U = np.zeros((n, n))

#definindo LDU
for i in range(n):
    for j in range(n):
        # Matriz triangular inferior
        if i > j:
            L[i, j] = A[i, j]
        # Matriz diagonal
        elif i == j:
            D[i, j] = A[i, j]
        # Matriz triangular superior
        else:
            U[i, j] = A[i, j]

#chute inicial
x = np.zeros([n])
erro = 0

for i in range(itMax):
    xNovo = np.linalg.inv(D) @ (b - (L + U) @ x)
    erro = np.linalg.norm(xNovo - x)
    x = xNovo
    
    if erro < epslon:
        break

print(f"iterações: {i}")
print(f"solução: {x}")
```

#### Método de Gauss-Seidel

```python
A = np.array([
    [5.0, 1.0, 1.0],
    [3.0, 4.0, 1.0],
    [3.0, 3.0, 6.0]
])
b = np.array([5.0, 6.0, 0.0])

itMax = 10
epslon = 1e-2
n = A.shape[0]

L = np.zeros((n, n))
U = np.zeros((n, n))
I = np.eye((n))

#definindo L* e U*
for i in range(n):
    for j in range(n):
        # Matriz triangular inferior
        if i > j:
            L[i, j] = A[i, j] / A[i, i]
        # Matriz triangular infoerior
        elif i < j:
            U[i, j] = A[i, j] / A[i, i]

for i in range(n):
    b[i] = b[i] / A[i, i]

#chute inicial
x = np.zeros([n])
erro = 0

for i in range(itMax):
    xNovo = - np.linalg.inv((L + I)) @ U @ x + np.linalg.inv((L + I)) @ b
    erro = np.linalg.norm(xNovo - x)
    x = xNovo

    if erro < epslon:
        break

print(f"iterações: {i}")
print(f"solução: {x}")
```

#### Método de SOR

```python
A = np.array([
    [5.0, 1.0, 1.0],
    [3.0, 4.0, 1.0],
    [3.0, 3.0, 6.0]
])
b = np.array([5.0, 6.0, 0.0])

itMax = 10
epslon = 1e-2
n = A.shape[0]
omega = 0.8

L = np.zeros((n, n))
U = np.zeros((n, n))
I = np.eye((n))

#definindo L* e U*
for i in range(n):
    for j in range(n):
        # Matriz triangular inferior
        if i > j:
            L[i, j] = A[i, j] / A[i, i]
        # Matriz triangular infoerior
        elif i < j:
            U[i, j] = A[i, j] / A[i, i]

for i in range(n):
    b[i] = b[i] / A[i, i]

#construindo A*
A = L + U + I

#chute inicial
xSOR = np.zeros(n)
xNovoSOR = np.zeros(n)
xGS = np.zeros(n)
erro = 0

for i in range(itMax):
    xNovoGS = - np.linalg.inv((L + I)) @ U @ x + np.linalg.inv((L + I)) @ b
    xGS = xNovoGS.copy()
    
    for j in range(n):
        soma1 = 0
        for k in range(j):
            soma1 += A[j, k] * xGS[k]
        soma2 = 0
        for k in range(j + 1, n):
            soma2 += A[j, k] * xSOR[k]
        xNovoSOR[j] = (1 - omega) * xSOR[j] + omega * (b[j] - soma1 - soma2)
    
    erro = np.linalg.norm(xNovoSOR - xSOR)
    xSOR = xNovoSOR.copy()

    if erro < epslon:
        break

print(f"iterações: {i}")
print(f"solução: {xSOR}")
```
