## Decomposição LU

```python
A = np.array([
    [5.0, 2.0, 1.0],
    [3.0, 1.0, 4.0],
    [1.0, 1.0, 3.0]
])
b = np.array([0.0, -7.0, -5.0])
n = A.shape[0]
U = np.zeros((n, n))
L = np.eye(n)

for i in range(n):

    for j in range(i, n):
        soma = 0

        for k in range(i):
            soma += L[i, k] * U[k, j]

        U[i, j] = A[i, j] - soma

    for j in range(i + 1, n):
        soma = 0

        for k in range(i):
            soma += L[j, k] * U[k, i]

        L[j, i] = (A[j, i] - soma) / U[i, i]

y = np.linalg.inv(L) @ b
x = np.linalg.inv(U) @ y

print(x)
```

## Decomposição de Cholesky

```python
A = np.array([
    [4.0, 2.0, -4.0],
    [2.0, 10.0, 4.0],
    [-4.0, 4.0, 9.0]
])
b = np.array([0.0, 6.0, 5.0])
n = A.shape[0]
G = np.zeros((n, n))

for i in range(n):
    for j in range(i + 1):
        if i == j:
            soma = 0
            for k in range(j):
                soma += G[i, k] ** 2
            G[i, j] = np.sqrt(A[i, j] - soma)
        else:
            soma = 0
            for k in range(j):
                soma += G[i, k] * G[j, k]
            G[i, j] = (A[i, j] - soma) / G[j, j]

y = np.linalg.inv(G) @ b
x = np.linalg.inv(G.T) @ y

print(x)
```
