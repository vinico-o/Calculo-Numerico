```python
import math

#lista dos valores
aproximacao = []
iteracoes = []
erros = []
metodos = ["bisseccao", "posicao falsa", "ponto fixo", "newton", "secante"]
```

## Método da Bissecção

```python
#intervalo
a = 0
b = 1

#precisao
e = 1e-3
erro = 1

def f(x):
    y = x * math.exp(-x) - math.exp(-3)
    return y

#iteracao atual
i = 1
while 1:
    x = (a + b) / 2
    erro = abs(b - a) / max(1, abs(b))

    if erro < e:
        break
    elif (f(a) * f(x) < 0):
        b = x;
    else:
        a = x;

    i += 1
    
print("Raiz aproximada:", x)
print("Iterações:", i)
print("Erro:", erro)

aproximacao.append(x)
iteracoes.append(i)
erros.append(erro)
```

# Método da Posição Falsa

```python
#intervalo
a = 0
b = 1

#precisao
e = 1e-3
erro = 1

def f(x):
    y = x * math.exp(-x) - math.exp(-3) 
    return y

#iteracao atual
i = 1
while 1:
    x = (a * f(b) - b * f(a)) / (f(b) - f(a))
    erro = abs(b - a) / max(1, abs(b))

    if erro < e:
        break
    elif (f(a) * f(x) < 0):
        b = x;
    else:
        a = x;

    i += 1
    
print("Raiz aproximada:", x)
print("Iterações:", i)
print("Erro:", erro)

aproximacao.append(x)
iteracoes.append(i)
erros.append(erro)
```

# Método Ponto Fixo

```python
def f(x):
    y = x * math.exp(-x) - math.exp(-3) 
    return y

def phi(x):
    y = math.exp(x) / math.exp(3)
    return y

#intervalo
a = 0
b = 1

#precisao
e = 1e-3
erro = 1

#iteracao atual
i = 1
#chute inicial
x = (b + a) / 2
while 1:
    x_antigo = x
    x = phi(x)
    erro = abs(x - x_antigo) / max(1, abs(x))

    if erro < e:
        break

    i += 1
    
print("Raiz aproximada:", x)
print("Iterações:", i)
print("Erro:", erro)

aproximacao.append(x)
iteracoes.append(i)
erros.append(erro)
```

## Método de Newton

```python
def f(x):
    y = x * math.exp(-x) - math.exp(-3)
    return y

#derivada de f(x)
def df(x):
    y =  math.exp(-x) - x * math.exp(-x)
    return y

#intervalo
a = 0
b = 1

#precisao
e = 1e-3
erro = 1

#iteracao atual
i = 1
#chute inicial
x = (b + a) / 2
while 1:
    x_antigo = x
    x = x_antigo - (f(x_antigo) / df(x_antigo))
    erro = abs(x - x_antigo) / max(1, abs(x))

    if erro < e:
        break

    i += 1
    
print("Raiz aproximada:", x)
print("Iterações:", i)
print("Erro:", erro)

aproximacao.append(x)
iteracoes.append(i)
erros.append(erro)
```

# Método da Secante

```python
def f(x):
    y = x * math.exp(-x) - math.exp(-3)
    return y

#intervalo
a = 0
b = 1

#precisao
e = 1e-3
erro = 1

#iteracao atual
i = 1
#chute inicial
x0 = a
x1 = b
while 1:
    x = x1 - ((x1 - x0) * f(x1)) / (f(x1) - f(x0))
    erro = abs(x - x1) / max(1, abs(x))

    if erro < e:
        break

    i += 1

    x0 = x1
    x1 = x

    
print("Raiz aproximada:", x)
print("Iterações:", i)
print("Erro:", erro)

aproximacao.append(x)
iteracoes.append(i)
erros.append(erro)
```
